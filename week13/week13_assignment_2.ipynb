### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
# The following code reads the dataset and then we get the shape (rows, columns)
import pandas as pd
covid=pd.read_csv('COVID-19_Cases.csv')
covid.shape
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?
```python
# We can examine our data with the describe function in pandas with a parameter of include = 'all' to show non-numerical stats
data_audit=pd.DataFrame(covid.describe(include='all')).T
data_audit['number of unique values']= covid.nunique()
data_audit.sort_values(['number of unique values'], ascending = False)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
dtype=pd.DataFrame(covid.dtypes .sort_values())
numeric_fields = dtype[dtype[0]!='object'].index
covid.hist(column=numeric_fields, figsize=(10,10))
```
